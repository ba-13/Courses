```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pickle
```

```python
file_path = "./mnist_small.pkl"

with open(file_path, 'rb') as file:
    data = pickle.load(file)
```

```python
X = data["X"]
Y = data["Y"]
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=Y, cmap='magma')
cbar = plt.colorbar(scatter, label='Label Value')
plt.title('PCA Projection of MNIST Digits')
plt.show()
```

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=Y, cmap='magma')
cbar = plt.colorbar(scatter, label='Label Value')
plt.title('t-SNE Projection of MNIST Digits')
plt.show()
```
